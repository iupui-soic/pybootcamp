```python

df_admissions = pd.read_csv('/content/gdrive/MyDrive/mimic-iv/admissions.csv.gz')
df_admissions
```

```python
col = ['admission_type','admission_location','insurance','language','race','marital_status']
plt.figure(figsize=(30,30))
for i, col in enumerate(col):
    axes = plt.subplot(5,2,i + 1)
    axes.pie(df_admissions[col].value_counts(), labels=df_admissions[col].value_counts().index, autopct='%.2f%%')
    axes.set_title(col)
plt.tight_layout()
plt.legend()
plt.show()
```
