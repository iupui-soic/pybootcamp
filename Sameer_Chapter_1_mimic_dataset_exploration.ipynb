```python
# Verify the columns in df_merge
print("Columns in df_merge before join operations:")
print(df_merge.columns)

# Load necessary DataFrames
df_procedures_icd = pd.read_parquet('/content/drive/MyDrive/mimiv-iv/procedures_icd.parquet')

# Ensure subject_id and hadm_id exist in df_merge before joining
if 'subject_id' in df_merge.columns and 'hadm_id' in df_merge.columns:
    df_merge = df_merge.join(df_procedures_icd.set_index(['subject_id', 'hadm_id']), on=['subject_id', 'hadm_id'], how='left')
    print(len(df_merge))
    print(df_merge.head())
else:
    print("Columns 'subject_id' and 'hadm_id' not found in df_merge")
```

```python
# Load necessary DataFrames
df_admissions = pd.read_parquet('/content/drive/MyDrive/mimiv-iv/admissions.parquet')
df_patients = pd.read_parquet('/content/drive/MyDrive/mimiv-iv/patients.parquet')
df_patients = df_patients.drop(['dod'], axis=1)
df_transfers = pd.read_parquet('/content/drive/MyDrive/mimiv-iv/transfers.parquet')
df_procedures_icd = pd.read_parquet('/content/drive/MyDrive/mimiv-iv/procedures_icd.parquet')

# Ensure the necessary columns are present in the initial DataFrames
print("Columns in df_admissions:")
print(df_admissions.columns)

print("Columns in df_patients:")
print(df_patients.columns)

print("Columns in df_transfers:")
print(df_transfers.columns)

# Perform initial merge operations, ensuring 'subject_id' and 'hadm_id' are included
df_merge = pd.merge(df_admissions, df_patients, on='subject_id')
print(f"Merged Admissions and Patients: {len(df_merge)}")
print(df_merge.head())

df_merge = pd.merge(df_merge, df_transfers, on=['subject_id', 'hadm_id'])
print(f"Merged with Transfers: {len(df_merge)}")
print(df_merge.head())

# Verify the columns in df_merge after initial merges
print("Columns in df_merge after initial merges:")
print(df_merge.columns)

# Perform join operations if 'subject_id' and 'hadm_id' are present
if 'subject_id' in df_merge.columns and 'hadm_id' in df_merge.columns:
    df_merge = df_merge.join(df_procedures_icd.set_index(['subject_id', 'hadm_id']), on=['subject_id', 'hadm_id'], how='left')
    print(len(df_merge))
    print(df_merge.head())
else:
    print("Columns 'subject_id' and 'hadm_id' not found in df_merge")
```
