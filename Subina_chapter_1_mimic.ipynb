```python
import pandas as pd
from mimic.mimic import MIMIC, MIMICManipulations

# Initialize MIMIC and MIMICManipulations instances
mimic_instance = MIMIC('/content/drive/MyDrive/mimiv-iv')
mimic_manipulation_instance = MIMICManipulations('/content/drive/MyDrive/mimiv-iv')

# Reading and filtering data from Parquet files
df_admission = mimic_manipulation_instance.filter_core_admission(pd.read_parquet('/content/drive/MyDrive/mimiv-iv/admissions.parquet'))
print(len(df_admission))
print(df_admission.head())

df_patient = mimic_manipulation_instance.filter_core_patients(pd.read_parquet('/content/drive/MyDrive/mimiv-iv/patients.parquet'))
print(len(df_patient))
print(df_patient.head())
```

```python
df_drgcodes = mimic_manipulation_instance.filter_hosp_drgcodes(pd.read_parquet('/content/drive/MyDrive/mimiv-iv/drgcodes.parquet'))
print(len(df_drgcodes))
print(df_drgcodes.head())
```

```python
df_merge = df_admission.merge(df_patient, left_index=True, right_index=True)
print(len(df_merge))
df_merge.head()
```

```python
df_admissions = pd.read_parquet('/content/drive/MyDrive/mimiv-iv/admissions.parquet')
df_patients = pd.read_parquet('/content/drive/MyDrive/mimiv-iv/patients.parquet')
```

```python
df_merge = pd.merge(df_admissions, df_patients, on='subject_id')
print(f"Merged Admissions and Patients: {len(df_merge)}")
print(df_merge.head())
```
