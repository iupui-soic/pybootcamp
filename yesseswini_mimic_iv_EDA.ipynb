```python

import zipfile
import os

# Define the base path where your zip files are located
base_path = '/content/drive/My Drive'
zip_files = ['dataset 1.zip', 'dataset 3.zip', 'dataset 4.zip', 'dataset 5.zip']

# Define the directory to extract the files
extract_dir = os.path.join(base_path, 'extracted_contents')

# Create the directory if it doesn't exist
if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)

# Unzip each dataset
for zip_file in zip_files:
    with zipfile.ZipFile(os.path.join(base_path, zip_file), 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
        print(f"Extracted {zip_file}")
```

```python
import os
import pandas as pd

# Define the path to the extracted directory
extract_dir = os.path.join('/content/drive/My Drive', 'extracted_contents', 'mimiv-iv')

# List all files in the `mimiv-iv` directory
extracted_files = os.listdir(extract_dir)
print("Extracted files in 'mimiv-iv':")
print(extracted_files)

# Initialize variables to store statistics
age_stats = None
gender_counts = None

# Iterate over the extracted files
for file in extracted_files:
    file_path = os.path.join(extract_dir, file)

    # Check if the file is a Parquet file for patients data
    if file == 'patients.parquet':
        try:
            # Read the Parquet file into a DataFrame
            patients_df = pd.read_parquet(file_path)

            # Calculate basic statistics for anchor_age
            age_stats = patients_df['anchor_age'].describe()

            # Count number of male and female patients
            gender_counts = patients_df['gender'].value_counts()

        except Exception as e:
            print(f"Error reading {file}: {e}")

# Print summary statistics for anchor_age
if age_stats is not None:
    print("\nSummary Statistics for anchor_age:")
    print(age_stats)

# Print number of patients by gender
if gender_counts is not None:
    print("\nNumber of patients by gender:")
    print(gender_counts)
```

```python
import os
import pandas as pd

# Define the path to the extracted directory
extract_dir = os.path.join('/content/drive/My Drive', 'extracted_contents', 'mimiv-iv')

# List all files in the `mimiv-iv` directory
extracted_files = os.listdir(extract_dir)
print("Extracted files in 'mimiv-iv':")
print(extracted_files)

# Function to read and preview data
def read_and_preview_data(file_path):
    try:
        if file_path.endswith('.parquet'):
            df = pd.read_parquet(file_path)
        elif file_path.endswith('.csv'):
            df = pd.read_csv(file_path)
        else:
            return  # Skip files that are neither Parquet nor CSV

        print(f"\nColumns in {os.path.basename(file_path)}:")
        print(df.columns.tolist())
        print(f"Preview of {os.path.basename(file_path)}:")
        print(df.head())  # Display the first few rows
        return df

    except Exception as e:
        print(f"Could not read {file_path}: {e}")
        return None

# Preview the contents of each dataset
for file in extracted_files:
    file_path = os.path.join(extract_dir, file)

    # Check if the file is a Parquet file related to microbiology events
    if 'microbiologyevents' in file and file.endswith('.parquet'):
        df_microbiology = read_and_preview_data(file_path)

        # Analyze Specimen Types
        if df_microbiology is not None and 'spec_type_desc' in df_microbiology.columns:
            print("\nSpecimen Types:")
            spec_type_counts = df_microbiology['spec_type_desc'].value_counts()
            print(spec_type_counts)

        # Analyze Organisms
        if df_microbiology is not None and 'org_name' in df_microbiology.columns and 'interpretation' in df_microbiology.columns:
            print("\nOrganisms:")
            organisms_info = df_microbiology[['org_name', 'interpretation']]
            print(organisms_info.head())

```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the path to the extracted directory
extract_dir = os.path.join('/content/drive/My Drive', 'extracted_contents', 'mimiv-iv')

# Function to read and preview data
def read_and_preview_data(file_path):
    try:
        if file_path.endswith('transfers.parquet'):
            df = pd.read_parquet(file_path)
        elif file_path.endswith('.csv'):
            df = pd.read_csv(file_path)
        else:
            return None

        print(f"\nColumns in {os.path.basename(file_path)}:")
        print(df.columns.tolist())
        print(f"Preview of {os.path.basename(file_path)}:")
        print(df.head())  # Display the first few rows
        return df

    except Exception as e:
        print(f"Could not read {file_path}: {e}")
        return None

# List all files in the `mimiv-iv` directory
extracted_files = os.listdir(extract_dir)
print("Extracted files in 'mimiv-iv':")
print(extracted_files)

# Analyze patient movement (careunit) over time using intime and outtime from transfers.parquet
for file in extracted_files:
    file_path = os.path.join(extract_dir, file)

    # Check if the file is transfers.parquet
    if file.endswith('transfers.parquet'):
        df_transfers = read_and_preview_data(file_path)

        # Care Unit Transfers Analysis
        if df_transfers is not None:
            if 'intime' in df_transfers.columns and 'outtime' in df_transfers.columns and 'careunit' in df_transfers.columns:
                try:
                    # Convert intime and outtime to datetime
                    df_transfers['intime'] = pd.to_datetime(df_transfers['intime'])
                    df_transfers['outtime'] = pd.to_datetime(df_transfers['outtime'])

                    # Group by careunit and year-month
                    df_transfers['year_month'] = df_transfers['intime'].dt.to_period('M')
                    transfer_counts = df_transfers.groupby(['year_month', 'careunit']).size().unstack().fillna(0)

                    # Plotting
                    plt.figure(figsize=(14, 8))

                    # Using seaborn for better aesthetics
                    sns.set_palette("colorblind")  # Set color palette
                    transfer_counts.plot(kind='line', marker='o', linestyle='-', ax=plt.gca())

                    plt.title('Care Unit Transfers Over Time')
                    plt.xlabel('Year-Month')
                    plt.ylabel('Transfer Count')
                    plt.grid(True)
                    plt.legend(title='Care Unit', fontsize='medium', shadow=True , bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=3)
                    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
                    plt.tight_layout()  # Adjust layout for better spacing

                    plt.show()

                except Exception as e:
                    print(f"Error analyzing care unit transfers from {os.path.basename(file_path)}: {e}")
```
