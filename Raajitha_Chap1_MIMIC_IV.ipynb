##Merge Data on the basis of subject_id index

1. Merging patients and admissions

```python
df_merge = df_admission.merge(df_patient, left_index=True, right_index=True)
print(len(df_merge))
df_merge.head()
```

2. Merging df_merge and df_transfers

```python
df_merge = df_merge.merge(df_transfers, left_index=True, right_index=True)
print(len(df_merge))
df_merge.head()
```

3. Merging df_merge and df_diagnoses_icd

```python
df_merge = df_merge.join(df_diagnoses_icd, how='left')
print(len(df_merge))
df_merge.head()
```

4. Merging df_merge and df_procedures_icd

```python
df_merge = df_merge.join(df_procedures_icd, how='left')
print(len(df_merge))
df_merge.head()
```

5. Merging df_merge with df_lab_events

6. Merging df_merge with df_drgcodes

```python
df_merge = df_merge.join(df_drgcodes, how='left')
print(len(df_merge))
df_merge.head()
```

7. Merging df_merge and df_emar

```python
df_merge= df_merge.join(df_emar, how='left')
print(len(df_merge))
df_merge.head()
```

8. Merging df_merge and df_prescriptions

```python
df_merge = df_merge.join(df_prescriptions, how='left')
print(len(df_merge))
df_merge.head()
```

9. Merging df_merge and df_microbiology_events

```python
df_merge = df_merge.join(df_microbiology_events, how='left')
print(len(df_merge))
df_merge.head()
```

10. Merging df_merge and df_services

```python
df_merge = df_merge.join(df_services, how='left')
print(len(df_merge))
df_merge.head()
```
